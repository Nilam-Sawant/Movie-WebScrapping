```python
import requests
from bs4 import BeautifulSoup
import lxml
import pandas as pd
```

```python
base='https://www.themoviedb.org'
base_url='https://www.themoviedb.org/movie?page='
url_lst = []
new_link = []
lst1 = []

# To fetch Main Website
response = requests.get(base).text
converted_data = BeautifulSoup(response,'lxml')

# To get all multiple Web Pages
for i in range(1,101):
    url_lst.append(base_url+str(i))

# To get all inner Web Pages
for val in url_lst:
    response = requests.get(val,timeout=30).text
    converted_data = BeautifulSoup(response,'lxml')
    all_page_div = converted_data.find_all('div',class_="card style_1")
    for item in all_page_div:
        link = item.find('a')['href']
        link1 = base+link
        response1 = requests.get(link1).text
        Converted_data1 = BeautifulSoup(response1, 'lxml')
        
        # movie_names
        movie_name = Converted_data1.find("h2")
        if movie_name:
            m_name = movie_name.text.strip()
        else:
            m_name = 'None'
        
        # Release Dates
        release_date = Converted_data1.find('span',class_ ='release')
        if release_date:
            r_date = release_date.text.strip()
        else:
            r_date = 'None'         
        
        # Movie Ratings
        movie_ratings = Converted_data1.find('div',class_='user_score_chart')
        if movie_ratings:
            m_ratings = Converted_data1.find('div',class_='user_score_chart')['data-percent']
        else:
            m_ratings = 'None'
        
        # Movie Genres
        movie_genres = Converted_data1.find('span',class_ ='genres')
        if movie_genres:
            m_genres=movie_genres.text.strip()
        else:
            m_genres = 'None'
        
        # Movie Duration
        movie_duration = Converted_data1.find('span',class_ ='runtime')
        if movie_duration:
            duration = movie_duration.text.strip()
        else:
            duration = 'None'
        
        # Directors Names
        director_name = Converted_data1.find_all('li',class_='profile')
        directors_lst=[]
        for data in director_name:
            if (data.find('p',class_='character').text.startswith('Director')):
                name=data.find('p').text
                directors_lst.append(name)
        
        # To get all details in Dictionary
        page_1_dict = {
            "Movie-Name":m_name,
            "Release-Date":r_date,
            "Movie Duration":duration,
            "Movie Ratings":m_ratings,
            "Movie Director":directors_lst,
            "Movie Genres":m_genres
            }

        # Converting in One List
        lst1.append(page_1_dict)
```

```python
df = pd.DataFrame(lst1)
df
```
